```python
# 15. Write a program to calculate the factorial of a number provided by the user using a while loop.

number = int(input("Enter a number: "))
factorial = 1
current = 1

while current <= number:
   factorial *= current
   current += 1

print(f"The factorial of {number} is {factorial}")
```
